<h2>
part 2 - print certificate
</h2>

```python
import OpenSSL
import ssl
```

```python
cert = open('C:\\Users\\Reza\\cert.pem','r').read()
x509 = OpenSSL.crypto.load_certificate(OpenSSL.crypto.FILETYPE_PEM, cert)
```

```python
issuer = x509.get_issuer()
subject = x509.get_subject()
valid_from = x509.get_notBefore().decode("utf-8") 
valid_from = valid_from[:8]
valid_from = valid_from[0:4]+'-'+valid_from[4:6]+'-'+valid_from[6:8]

valid_to = x509.get_notAfter().decode("utf-8") 
valid_to = valid_to[:8]
valid_to = valid_to[0:4]+'-'+valid_to[4:6]+'-'+valid_to[6:8]

print('Common name:{}'.format(issuer.CN))
print('valid from: {}'.format(valid_from))
print('valid to: {}'.format(valid_to))
print('Issuer name:{}'.format(subject.CN))
```
